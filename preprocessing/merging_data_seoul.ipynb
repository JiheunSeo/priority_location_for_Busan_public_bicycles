```python
import pandas as pd
import numpy as np

top100 = pd.read_csv('./data/읍면동별_자전거_총대여반납건수_상위100개_서울.csv',encoding='cp949')
top50 = pd.read_csv('./data/읍면동별_자전거_총대여반납건수_상위50개_서울.csv',encoding='cp949') #csv 파일 불러옴
```

```python
school = pd.read_csv('./data/읍면동별_학교_서울.csv',encoding='cp949') #행정동별 학교 개수 추출
school["학교개수"] = ""
school = school.groupby('행정동명').학교개수.count().to_frame()

movie = pd.read_csv('./data/읍면동별_영화관_서울.csv',encoding='cp949') #행정동별 영화관 개수 추출
movie["영화관개수"] = ""
movie = movie.groupby('행정동명').영화관개수.count().to_frame()

university = pd.read_csv('./data/읍면동별_대학교_서울.csv',encoding='cp949') #행정동별 대학교 개수 추출
university["대학교개수"] = ""
university = university.groupby('행정동명').대학교개수.count().to_frame()
```

```python
#변수 csv파일 불러오기
facility = pd.read_csv('./data/읍면동별_시설물수_서울.csv',encoding='cp949')
populationDensity = pd.read_csv('./data/읍면동별_인구밀도및인구수_서울.csv',encoding='UTF8')
river = pd.read_csv('./data/읍면동별_인접한강_서울.csv',encoding='cp949')
stream = pd.read_csv('./data/읍면동별_인접한하천_서울.csv',encoding='cp949')
car = pd.read_csv('./data/읍면동별_자동차등록대수_서울.csv',encoding='cp949')
byciclePath = pd.read_csv('./data/읍면동별_자전거도로_서울.csv',encoding='cp949')
subway = pd.read_csv('./data/읍면동별_지하철승객수_서울.csv',encoding='UTF8')
bus = pd.read_csv('./data/읍면동별_버스승객수_서울.csv',encoding='UTF8')
```

## 읍면동별_자전거_총대여반납건수_상위100개_서울

```python
#변수 join
data100=pd.merge(left = top100,right = facility,on='행정동_코드',how="left")
data100=pd.merge(left = data100,right = populationDensity,on="행정동명",how="left")
data100=pd.merge(left = data100, right = river,on="행정동명",how="left")
data100=pd.merge(left = data100, right = stream,on="행정동명",how="left")
data100=pd.merge(left = data100, right = car,on="행정동명",how="left")
data100=pd.merge(left = data100, right = byciclePath,on="행정동명",how="left")
data100=pd.merge(left = data100, right = subway,on='행정동_코드',how="left")
data100=pd.merge(left = data100, right = bus,on='행정동_코드',how="left")
data100=pd.merge(left = data100, right = school,on='행정동명',how="left")
data100=pd.merge(left = data100, right = movie,on='행정동명',how="left")
data100=pd.merge(left = data100, right = university,on='행정동명',how="left")
```

```python
#중복행 삭제
data100.drop_duplicates()
```

```python
data100.to_csv('./data/seoul_data100.csv',index=False)
```
