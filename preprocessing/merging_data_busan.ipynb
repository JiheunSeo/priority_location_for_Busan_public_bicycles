```python
import pandas as pd
import numpy as np
```

```python
school = pd.read_csv('./data/읍면동별_학교_부산.csv',encoding='cp949') #행정동별 학교 개수 추출
school["학교개수"] = ""
school = school.groupby('행정동명').학교개수.count().to_frame()
school.reset_index(inplace=True)
school.columns = ['행정동명', '학교개수']

movie = pd.read_csv('./data/읍면동별_영화관_부산.csv',encoding='cp949') #행정동별 영화관 개수 추출
movie["영화관개수"] = ""
movie = movie.groupby('행정동명').영화관개수.count().to_frame()
movie.reset_index(inplace=True)
movie.columns = ['행정동명', '영화관개수']

university = pd.read_csv('./data/읍면동별_대학교_부산.csv',encoding='cp949') #행정동별 대학교 개수 추출
university["대학교개수"] = ""
university = university.groupby('행정동명').대학교개수.count().to_frame()
university.reset_index(inplace=True)
university.columns = ['행정동명', '대학교개수']

facility = pd.read_csv('./data/읍면동별_시설물_부산.csv',encoding='cp949') #행정동별 대학교 개수 추출
facility["시설물개수"] = ""
facility = facility.groupby('행정동명').시설물개수.count().to_frame()
facility.reset_index(inplace=True)
facility.columns = ['행정동명', '시설물개수']
```

```python
#변수 csv파일 불러오기
populationDensity = pd.read_csv('./data/읍면동별_인구밀도_부산.csv',encoding='cp949')
river = pd.read_csv('./data/읍면동별_인접한강_부산.csv',encoding='cp949')
stream = pd.read_csv('./data/읍면동별_인접한하천_부산.csv',encoding='cp949')
car = pd.read_csv('./data/읍면동별_자동차등록대수_부산.csv',encoding='cp949')
byciclePath = pd.read_csv('./data/읍면동별_자전거도로_부산.csv',encoding='cp949')
subway = pd.read_csv('./data/읍면동별_지하철승객수_부산.csv',encoding='cp949')
#bus = pd.read_csv('./data/읍면동별_버스승객수_부산.csv',encoding='UTF8')
```

## 읍면동별_자전거_총대여반납건수_부산

```python
#변수 join
data=pd.merge(left = populationDensity ,right = facility,on='행정동명',how="left")
data=pd.merge(left = data,right = car,on="행정동명",how="left")
data=pd.merge(left = data, right = river,on="행정동명",how="left")
data=pd.merge(left = data, right = stream,on="행정동명",how="left")
data=pd.merge(left = data, right = byciclePath,on="행정동명",how="left")
data=pd.merge(left = data, right = subway,on='행정동명',how="left")
data=pd.merge(left = data, right = school,on='행정동명',how="left")
data=pd.merge(left = data, right = movie,on='행정동명',how="left")
data=pd.merge(left = data, right = university,on='행정동명',how="left")
#data=pd.merge(left = data, right = bus,on='행정동_코드',how="left")
```

```python
#중복행 삭제
data.drop_duplicates()
```

```python
data.to_csv('./data/busan_data.csv',index=False, encoding = 'cp949')
```
