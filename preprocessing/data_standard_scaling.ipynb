```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
seoul_data = pd.read_csv('./data/seoul_data100.csv',encoding='cp949')
busan_data = pd.read_csv('./data/busan_data.csv',encoding='cp949')
```

```python
seoul_data = seoul_data.drop('인구 (명)', axis = 1)
seoul_data = seoul_data.drop('면적 (㎢)', axis = 1)
seoul_data = seoul_data.drop('구', axis = 1)
seoul_data = seoul_data.drop('행정동_코드', axis = 1)
seoul_data = seoul_data.drop('총반납대여', axis = 1)
seoul_data = seoul_data.drop('버스_총승객수', axis = 1)
seoul_dong = seoul_data.iloc[:,[0]]
seoul_data = seoul_data.drop('행정동명', axis = 1)

busan_dong = busan_data.iloc[:,[0]]
busan_data = busan_data.drop('행정동명', axis = 1)
```

```python
seoul_data
```

```python
seoul_data = pd.get_dummies(seoul_data, columns = ['강'])
seoul_data['하천'] = seoul_data['하천'].fillna(0).map(lambda x: 1 if x != 0 else x)
seoul_data = seoul_data.fillna(0)

busan_data = pd.get_dummies(busan_data, columns = ['강'])
busan_data['하천'] = busan_data['하천'].fillna(0).map(lambda x: 1 if x != 0 else x)
busan_data = busan_data.fillna(0)
```

```python
array_seoul_data = seoul_data.to_numpy() #array꼴로 바꿈
array_busan_data = busan_data.to_numpy() #array꼴로 바꿈
```

```python
busan_data
```

```python
seoul_data
```

```python
seoul_data_scaled = scaler.fit_transform(array_seoul_data)
busan_data_scaled = scaler.fit_transform(array_busan_data)
```

```python
seoul_df_scaled = pd.DataFrame(seoul_data_scaled, columns=['시설물개수', '인구밀도', '하천', '자동차등록대수','자전거도로','지하철_총승객수','학교개수', '영화관개수', '대학교개수' ,'강'])
busan_df_scaled = pd.DataFrame(busan_data_scaled, columns=['인구밀도', '시설물개수', '자동차등록대수', '하천','자전거도로','지하철_총승객수','학교개수', '영화관개수', '대학교개수', '강'])
```

```python
busan_df_scaled = busan_df_scaled[['하천', '자동차등록대수'] + [col for col in busan_df_scaled.columns if col not in ['자동차등록대수', '하천']]]
busan_df_scaled = busan_df_scaled[['시설물개수', '인구밀도'] + [col for col in busan_df_scaled.columns if col not in ['인구밀도', '시설물개수']]]
```

```python
busan_df_scaled['행정동명'] = busan_dong
seoul_df_scaled['행정동명'] = seoul_dong
```

```python
seoul_data = pd.read_csv('./data/seoul_data100.csv',encoding='UTF8')
```

```python
seoul_data = pd.read_csv('./data/읍면동별_자전거_총대여반납건수_상위100개_서울.csv',encoding='cp949')
seoul_data
```

```python
seoul_dong = seoul_data['행정동명'] #서울동저장
seoul_data = seoul_data.drop('행정동명', axis = 1)

seoul_cluster = seoul_data['cluster'] #서울클러스터저장
seoul_data = seoul_data.drop('cluster', axis = 1)

desired_min = 0
desired_max = 2

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# MinMaxScaler를 사용하여 스케일링
scaler = MinMaxScaler(feature_range=(desired_min, desired_max))
data_scaled = scaler.fit_transform(seoul_data)


seoul_data_scaled = pd.DataFrame(data_scaled, columns=['행정동코드','총반납대여'])


seoul_data_scaled['행정동명'] = seoul_dong
seoul_data_scaled['cluster'] = seoul_cluster
```

```python
seoul_data_scaled.to_csv("./data/seoul_minmax_newdata0to2.csv", index=False, encoding = 'cp949')
```
